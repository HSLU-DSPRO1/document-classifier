```
# Optional: convert PDFs to page images for vision inference
# Requires: pymupdf
import fitz
from PIL import Image
import numpy as np

PDF_DIR = INPUT_DIR / "pdfs"
OUT_IMG_DIR = ROOT / "outputs" / "pdf_pages"
OUT_IMG_DIR.mkdir(parents=True, exist_ok=True)

def pdf_to_images(pdf_path: Path, max_pages: int = 2, dpi: int = 144):
    doc = fitz.open(pdf_path)
    images = []
    try:
        n = min(doc.page_count, max_pages)
        for i in range(n):
            page = doc.load_page(i)
            pix = page.get_pixmap(dpi=dpi)
            img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
            images.append((i, img))
    finally:
        doc.close()
    return images

# example:
pdfs = sorted(PDF_DIR.glob("*.pdf")) if PDF_DIR.exists() else []
pdfs[:3]
```
